```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
adam = pd.read_csv('logs/CIFAR-10_Pretrain_adam.csv')
ga = pd.read_csv('logs/CIFAR-10_HYBRID_GA.csv')
meme = pd.read_csv('logs/CIFAR-10_meme_steady_state.csv')
```

```python
# Plot

fig ,(ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

ax1.plot(adam['train_acc'], label='Adam')
ax1.plot(ga['train_acc'], label='GA')
ax1.plot(meme['train_acc'], label='Chosen Optimizer')
ax1.set_title("Accuracy on Training")
ax1.set_ylabel('Accuracy')
ax1.set_xlabel('Iterations (Epochs or Generations)')
ax1.legend(loc="lower right")


ax2.plot(adam['train_loss'], label="Adam")
ax2.plot(ga['train_loss'], label="GA")
ax2.plot(meme['train_loss'], label="Chosen Optimizer")
ax2.set_title("Loss on Training")
ax2.set_ylabel('Loss')
ax2.set_xlabel('Iterations (Epochs or Generations)')
ax2.legend(loc="upper right");
```
