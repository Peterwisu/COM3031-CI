```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Gradient Descent

###  Adam optimizer with pretrain CNN

```python
adam = pd.read_csv('logs/CIFAR-10_Pretrain_adam.csv')

fig ,(ax1) = plt.subplots(nrows=1, ncols=1, figsize=(5,4))

ax1.plot(adam['train_acc'], label='Adam accuracy',linestyle='--', color='b')

ax1.set_title("Accuracy on Training")
ax1.set_ylabel('Accuracy')
ax1.set_xlabel('Iterations (Epochs)')
ax1.legend(loc="upper left")
ax1.grid()
```

## Population-based meta-heuristic algorithms

### Generational vs Steady State

```python
# Plot comparision of result of Genetic Algorithms with different Survival Selection
ga_generational = pd.read_csv('logs/HYBRID_GA_real_coded_generational.csv')
ga_steady_state = pd.read_csv('logs/HYBRID_GA_real_coded_steady_state.csv')

fig ,(ax1) = plt.subplots(nrows=1, ncols=1, figsize=(5,4))
ax1.plot(ga_generational['train_acc'], label='Generational',linestyle='--', color='r')
ax1.plot(ga_steady_state['train_acc'], label='Steady State',linestyle='--', color='b')

ax1.set_title("Accuracy on Training")
ax1.set_ylabel('Accuracy')
ax1.set_xlabel('Iterations (Epochs or Generations)')
ax1.legend(loc="upper left")
ax1.grid()
```

### Compare GA(Binary and Real encoding) and PSO

```python
# Plot comparison on population based optimizer
ga_bc = pd.read_csv('logs/HYBRID_GA_binary_coding_steady_state.csv')
ga_real = pd.read_csv('logs/HYBRID_GA_real_coded_steady_state.csv')
pso = pd.read_csv('logs/HYBRID_PSO_GLOBAL_100_SWARM.csv')

fig ,(ax1) = plt.subplots(nrows=1, ncols=1, figsize=(5,4))
ax1.plot(ga_bc['train_acc'], label='GA with Binary coding',linestyle='--', color='b')
ax1.plot(ga_real['train_acc'], label='GA with real coding',linestyle='--', color='r')
ax1.plot(pso['train_acc'], label='PSO',linestyle='--', color='g')
ax1.set_title("Accuracy on Training")
ax1.set_ylabel('Accuracy')
ax1.set_xlabel('Iterations (Epochs or Generations)')
ax1.legend(loc="lower right")
ax1.grid()
```

## Memetic Algorithms

### Encoding in Memetic Algorithms

```python
# compare a result of Binary encoding and real coded in memetic algorithms
meme_bc = pd.read_csv('logs/memetic-rprop-binary.csv')
meme_real = pd.read_csv('logs/memetic-rprop-realcode.csv')

fig ,(ax1) = plt.subplots(nrows=1, ncols=1, figsize=(5,4))
ax1.plot(meme_bc['train_acc'], label='Memetic with Binary coding',linestyle='--', color='b')
ax1.plot(meme_real['train_acc'], label='Memetic with real coded',linestyle='--', color='r')

ax1.set_title("Accuracy on Training")
ax1.set_ylabel('Accuracy')
ax1.set_xlabel('Iterations (Epochs or Generations)')
ax1.legend(loc="lower right")
ax1.grid()
```

### Compare Three optimizer 

```python
# Plot comparision of all optimizer

adam = pd.read_csv('logs/CIFAR-10_Pretrain_adam.csv')
ga = pd.read_csv('logs/HYBRID_GA_real_coded_steady_state.csv')
meme = pd.read_csv('logs/memetic-adam-realcode_with_5_iteration_with50pop.csv')

fig ,(ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

ax1.plot(adam['train_acc'], label='Adam', linestyle='--', color='b')
ax1.plot(ga['train_acc'], label='GA', linestyle='--', color='r')
ax1.plot(meme['train_acc'], label='Chosen Optimizer',linestyle='--', color='g')
ax1.set_title("Accuracy on Training")
ax1.set_ylabel('Accuracy')
ax1.set_xlabel('Iterations (Epochs or Generations)')
ax1.legend(loc="lower right")
ax1.grid()


ax2.plot(adam['train_loss'], label="Adam",linestyle='--', color='b')
ax2.plot(ga['train_loss'], label="GA",linestyle='--', color='r')
ax2.plot(meme['train_loss'], label="Chosen Optimizer",linestyle='--', color='g')
ax2.set_title("Loss on Training")
ax2.set_ylabel('Loss')
ax2.set_xlabel('Iterations (Epochs or Generations)')
ax2.legend(loc="upper right")
ax2.grid();
```
